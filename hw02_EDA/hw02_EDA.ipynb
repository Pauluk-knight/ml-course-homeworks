```python
import numpy as np
import pandas as pd
 

df = pd.read_csv('WineQT.csv')

print(f"Size of dataframe: {df.shape}")
display(df.head())
display(df.tail())
 
```

```python
print ("\n ----- Tecnichal information -------")
df.info()
df.describe()
display(df.describe())

print ("Count of пропусков по колонкам ")
print (df.isnull().sum())


dublicates_count = df.duplicated().sum()
print(f"Count of полных дубликатов: {dublicates_count}")
df = df.drop_duplicates()

print(f"\nData types:\n {df.dtypes}")
```

Нейронка предложила испортить мне мой датасет, а после героически его вылечить, чтобы выполнить каждый пункт дз. Я то в принципе не против, поэтому делаю

```python
df_dirty = df.copy()
# Добавим немного хаоса: занулим случайные ячейки в разных колонках
df_dirty.loc[df_dirty.sample(frac=0.05).index, 'fixed acidity'] = np.nan  # Числовой признак
df_dirty.loc[df_dirty.sample(frac=0.05).index, 'volatile acidity'] = np.nan # Еще числовой
df_dirty['wine_type'] = np.random.choice(['Red', 'White', np.nan], size=len(df_dirty)) # Категория

print(df_dirty.isnull().sum())
```

Заполнять пропуски следует средним значением, если данная категория не обладает серьезными выбросами и довольно симметрична. 

```python
mean_val = df_dirty['fixed acidity'].mean()
df_dirty['fixed acidity'] = df_dirty['fixed acidity'].fillna(mean_val)
print(f"Заполнили 'fixed acidity' средним: {mean_val:.2f}")
```

В случае,если в категории есть выбросы, которые критически влияют на среднее арифметическое значение, стоит использовать медианное данной категории

тут разница между средним и медианным 0,01, в силу того, что не так много было выбросов с максимальным значением, жалко(((

```python
median_val = df_dirty['volatile acidity'].median()
df_dirty['volatile acidity'] = df_dirty['volatile acidity'].fillna(median_val)
print(f"Заполним 'volatile acidity' медианным: {median_val:.2f}")
print (df_dirty['volatile acidity'].mean())
```

Заполнение модой скорее всего требуется именно там,где характеристика не сильно влияет на таргет значение, а где просто требуется самое часто-встречаемое слово.В нашем случае это может быть только название вина

```python
wine_mode = df_dirty['wine_type'].mode()[0]
print (f"Wine type mode is: {wine_mode}")
df_dirty['wine_type'] = df_dirty['wine_type'].fillna(wine_mode)
print(f"Заполнили 'wine_type' модой: {wine_mode}")
```

Если есть колонка,которая критически влияет на итоговый таргет, например качество самого вина, то тут лучше не гадать, ибо придуманный вариант заполнения может сильно ухудшить выдаваемый результат, так что лучше просто удалить эти пропуски


```python

df_dirty.dropna(subset=['quality'], inplace=True)
print("\nПроверка после очистки:")
print(df_dirty.isnull().sum())
```

 1. Базовая статистика + Перцентили
 .describe() уже дает 25, 50, 75. Добавим 5% и 95% для поиска экстремальных значений.

```python
cols_to_analyze = ['alcohol', 'pH', 'residual sugar']
stats = df[cols_to_analyze].describe(percentiles=[.05, .25, .5, .75, .95])

extra_metrics = {}

for col in cols_to_analyze:
    extra_metrics[col] = {
        'mode': df[col].mode()[0],            # Мода (самое частое)
        'variance': df[col].var(),            # Дисперсия
        'skewness': df[col].skew(),           # Асимметрия
        'kurtosis': df[col].kurtosis()         # Эксцесс
    }

extra_df = pd.DataFrame(extra_metrics)
full_stats = pd.concat([stats, extra_df])
display(full_stats)
```

Дисперсия --- среднй квадрат отклонения от среднего показателя.
Условно среднее количество сахара --- 2,52, но есть выбросы со значением сахара 0,9 , и со значением 15,5.
Т.е. чем больше дисперсия , тем  больше и резче случаи особых выбросов, т.е. вина есть как и сладкие, так и полусладкие


Ассиметрия - насколько график скошен в одну или другую сторону.Варируется в окрестности нуля. Если больше --- то скос вправо,значит больше вин с высоким показателем сахара, нежели низким,это отразится на нашем боксе с усами,думаю ус правый длинее будет и в случае сильных перепадов у полусладких вин, правая часть бокса тоже широкой будет

Эксцесс --- насколько острый или сглаженный пик у графика 
если высокий, то данные плотно сгрупированы относительно какой-то точки, низкий  --- данные распределены равномерно по всем видам вин


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['residual sugar'], color='skyblue', flierprops={"marker": "x"})
plt.title('Распределение остаточного сахара (Residual Sugar)', fontsize=15)
plt.xlabel('Сахар (г/дм³)', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

1) узкий ящик --- значит большинство все-таки (50%) вин со схожем содержанием сахара
2) Длинный правый ус и кучу выбросов особенных сладких вин (дессертные вина )
3) Ассиметрия --- немного заметно, что медиана смещена влево, положительная ассиметрия

```python
# 1. Отношение сахара к алкоголю (условная "сладость" восприятия)
df['sugar_to_alcohol'] = df['residual sugar'] / df['alcohol']

# 2. Общая кислотность (сумма фиксированной и летучей кислотности)
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']

# 3. Категориальный признак: тип качества (для энкодинга)
# Разделим вина на 'Low' (3-5), 'Medium' (6) и 'High' (7-8)
df['quality_label'] = pd.cut(df['quality'], 
                             bins=[0, 5, 6, 10], 
                             labels=['Low', 'Medium', 'High'])

print("Датафрейм после генерации новых фич:")
display(df[['quality', 'quality_label', 'sugar_to_alcohol', 'total_acidity']].head())
```

```python
df_encoded = pd.get_dummies(df, columns=['quality_label'], prefix='quality', drop_first=False)

print("\nДобавились колонки после One-Hot Encoding:")
display(df_encoded.filter(like='quality_').head())
```
